### ML Strategy - Key Points

- Define a single evaluation metric and potentially other satisfying metrics.
- Split data into training, validation and test sets.
  - Iterate on training and development sets to improve performance.
  - Dev and test sets should come from the same distribution and be representative of the data the model will see in production.
  - Use test set to evaluate performance.
  - If the data size is very large use a more skewed 98/1/1% split
- For some problems understanding `human level performance` is a good proxy for understanding the best possible performance (`Bayes optimal error`). With this, better error analysis can be done pointing out whether it is a bias or variance problem. `Avoidable bias` is the difference between the human level performance (what we want to achieve) and the training error. `Variance` is the difference between the training error and the development error.
- Look at individual examples and understand the causes of error.
- Break the examples into groups and look at the causes of error.
- Build your first system quickly, then iterate.
- When extra data is available but from a different distribution - it can help but it is best to leave the dev and test sets as they are as they are representative of the data the model will see in production.
- Transfer learning is a technique where a pre-trained model is used as a starting point for a new task perhaps training only the last layer or two.
- Multi-task learning is a technique where a single model is used to solve multiple related problems for example predicting whether a user will click on an ad or not and whether they will buy a product.
- End-to-end learning is a technique where the problem is approached as a single problem and a model is trained to solve it from start to finish.
